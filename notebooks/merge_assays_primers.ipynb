```python
import pandas as pd
```

```python
dfp = pd.read_csv('../primers_without_assay_info.tsv', sep='\t', index_col=False)
dft = pd.read_csv('../assays_without_primer_info.tsv', sep='\t', index_col=False)
```

```python
dfp
```

```python
# create a new dfp2 from dfp where the direction column is used to create new columns pcr_primer_forward, pcr_primer_reverse, pcr_primer_name_forward, pcr_primer_name_reverse, pcr_primer_reference_forward, pcr_primer_reference_reverse
dfp2 = dfp.pivot(index='pcr_primer_name', columns='direction', values=['pcr_primer', 'pcr_primer_name', 'pcr_primer_reference'])
dfp2.columns = ['_'.join(col).strip() for col in dfp2.columns.values]
dfp2 = dfp2.reset_index()
# Drop the pcr_primer_name column from dfp2
dfp2 = dfp2.drop(columns=['pcr_primer_name'])
```

```python
dfp2
```

```python
dft
```

```python
dft2 = dft.copy(deep=True)
# Ensure unique index for mapping
dfp2_fwd = dfp2.dropna(subset=['pcr_primer_forward'], inplace=False).set_index('pcr_primer_forward')
dfp2_rev = dfp2.dropna(subset=['pcr_primer_reverse'], inplace=False).set_index('pcr_primer_reverse')
# Map the new columns from dfp2 to dft2
dft2['pcr_primer_name_forward'] = dft['pcr_primer_forward'].map(dfp2_fwd['pcr_primer_name_forward'])
dft2['pcr_primer_name_reverse'] = dft['pcr_primer_reverse'].map(dfp2_rev['pcr_primer_name_reverse'])
dft2['pcr_primer_reference_forward'] = dft['pcr_primer_forward'].map(dfp2_fwd['pcr_primer_reference_forward'])
dft2['pcr_primer_reference_reverse'] = dft['pcr_primer_reverse'].map(dfp2_rev['pcr_primer_reference_reverse'])
```

```python
dft2
```

```python
dft2.to_csv('../assays.tsv', sep='\t', index=False)
```
